```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
import seaborn as sns
```

```python
# Load dataset from raw folder
df = pd.read_csv("../data/raw/house_prediction_data_set.csv", sep=r"\s+", header=None)


print("Shape:", df.shape)
print(df.columns.tolist())
print(df.head())
```

```python
#s used to rename the columns of dataset.
df.columns = [
    "CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE",
    "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"
]
```

```python
print(df.shape)
print(df.info())
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Relationship between RM (avg rooms) and MEDV
sns.scatterplot(x="RM", y="MEDV", data=df)
plt.title("RM vs MEDV")
plt.show()

# Relationship between LSTAT (% lower status) and MEDV
sns.scatterplot(x="LSTAT", y="MEDV", data=df)
plt.title("LSTAT vs MEDV")
plt.show()
```
