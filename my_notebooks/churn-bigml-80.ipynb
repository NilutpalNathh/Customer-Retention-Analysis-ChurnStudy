```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/churn-bigml-80.csv")
```

```python
print(df.head())
print(df.info())
```

```python
print("Missing values:\n", df.isnull().sum())

```

```python
#Drop duplicates
df = df.drop_duplicates()
print(df.head())
```

```python
#After dropping duplicates
print("Rows after:", len(df))
```

```python
#claind dataset

print(df.head())
```

```python
# Handle missing values
df = df.dropna()
print(df.head())
```

```python
#Standardize data types
if "Total day minutes" in df.columns:
    df["Total day minutes"] = df["Total day minutes"].astype(float)
if "Total intl calls" in df.columns:
    df["Total intl calls"] = df["Total intl calls"].astype(int)
    print(df.head())
```

```python
#Strip spaces from categorical columns (if any)
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].str.strip()
    print(df.head())
```

```python
#Final check
print("Final dataset info:")
print(df.info())
print("\nSample data:")
print(df.head())
```

```python
#Save cleaned dataset
df.to_csv("../data/processed/churn-bigml-80_cleaned.csv", index=False)
print("✅ Cleaned dataset saved as churn-bigml-80_cleaned.csv")
print(df.head())
```
