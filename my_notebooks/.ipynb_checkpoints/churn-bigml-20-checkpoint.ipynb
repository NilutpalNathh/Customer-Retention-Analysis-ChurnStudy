```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/churn-bigml-20.csv")
```

```python
print(df.head())
print(df.info())
print(df.shape)
print(df.describe())
```

```python
#Check for missing values:
print(df.isnull().sum())
```

```python
df.dropna()
```

```python
#Drop duplicates:
df = df.drop_duplicates()
print("After removing duplicates:", df.shape)
```

```python
## Clean categorical text columns
df["International plan"] = df["International plan"].str.strip().str.capitalize()
df["Voice mail plan"] = df["Voice mail plan"].str.strip().str.capitalize()

# Ensure 'Churn' is boolean
df["Churn"] = df["Churn"].astype(bool)

print(df[["International plan", "Voice mail plan", "Churn"]].head())
print(df.info())
```

```python
print("Final dataset info:")
print(df.info())
print("\nSample data:")
print(df.head())
```

```python
#Save the cleaned dataset
df.to_csv("../data/processed/churn_bigml_20_cleaned.csv", index=False)
print("Cleaned dataset saved successfully!")
```
