```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/house_prediction_data_set.csv", sep="\s+")
```

```python
print(df.head())
print(df.info())
```

```python
#Check column names
print(df.columns)
```

```python
#Check for missing values
print(df.isnull().sum())
```

```python
#Drop duplicates
df = df.drop_duplicates()
print(df.head())
```

```python
#Standardize data types
df = df.apply(pd.to_numeric, errors='coerce')
print(df.head())
```

```python
#handiling missing values
df = df.fillna(df.mean())

print(df.dtypes)
```

```python
#Save cleaned dataset
df.to_csv("../data/processed/house_prediction_data_cleaned.csv", index=False)
print("✅ File saved as house_prediction_data_cleaned.csv in ../data/processed/")
print(df.dtypes)
```
