```python
#Prepare for Machine Learning
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df["species_encoded"] = encoder.fit_transform(df["species"])
print(df.head())

```

```python
#Split features/labels:

X = df.drop(columns=["species", "species_encoded"])
y = df["species_encoded"]

print(X.head())
print(y.head())

```

```python
# Check Missing Values
```

```python
print(df.isnull().sum())
df = df.drop_duplicates() 
```

```python
#Check Duplicates
```

```python
print("Duplicates:", df.duplicated().sum())

```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print(df.columns)
#Standardize Column Names
```

```python
print(df.dtypes)
```

```python
#numeric columns are stored as strings, convert:
```

```python
df["sepal_length"] = pd.to_numeric(df["sepal_length"], errors="coerce")
```

```python
#Handle Inconsistent Categories
df["species"] = df["species"].str.strip().str.lower()
print(df["species"].unique())
```
